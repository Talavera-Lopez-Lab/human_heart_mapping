```python
import pandas as pd
```

```python
meta_data = pd.read_csv(
    filepath_or_buffer='../.data/meta_data/meta_phs002031.csv',
    sep='\t'
)
meta_data.head()
```

```python
meta_data = (
    meta_data
    .assign(sample=lambda x: x['file_name '].str.split('.').str[0],
            url= lambda x: 'https://cirm.ucsc.edu/cgi-bin/cdwGetFile?acc=' + x['sample'])
)
meta_data
```

```python
meta_data_download = meta_data[['file_name ', 'sample', 'url']]
meta_data_download.to_csv('../.data/meta_data/downloads_table.csv')
meta_data_download.head()
```
