```python
import pandas as pd
```

```python
rna_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12916.sdrf.txt',
                                  index_col='Description')
rna_seq_meta_data.head()
```

```python
atac_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12919.sdrf.txt',
                                   index_col='Description')
atac_seq_meta_data.head()
```

```python
keys = ['rna', 'atac']
meta_data = pd.concat(
    [rna_seq_meta_data, atac_seq_meta_data],
    axis=1,
    keys= keys
)
meta_data.head()
```

```python
columns = [
    'Source Name',
    'Comment[SUBMITTED_FILE_NAME]',
    'Comment[FASTQ_URI]'
]

meta_data_download = pd.concat([meta_data[key][columns].add_suffix(f'_{key}') for key in keys], axis=1)
meta_data_download.head()
```

```python
meta_data_download['Comment[FASTQ_URI]_rna'].iloc[0]
```

```python
meta_data_download.to_csv('../.data/downloads_table.csv')
```
