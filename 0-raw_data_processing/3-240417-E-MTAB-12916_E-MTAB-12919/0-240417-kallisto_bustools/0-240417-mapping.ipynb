```python
import pandas as pd
```

```python
rna_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12916.sdrf.txt',
                                  index_col='Description')
rna_seq_meta_data.head()
```

```python
atac_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12919.sdrf.txt',
                                   index_col='Description')
atac_seq_meta_data.head()
```

```python
meta_data = pd.concat(
    [rna_seq_meta_data, atac_seq_meta_data],
    axis=1,
    keys=['rna', 'atac']
)
meta_data.head()
```

```python
columns = {
    'Source Name',
    'Comment[ENA_SAMPLE]'
}
```

```python
pd.concat([meta_data[key][columns] for key in meta_data.keys()], axis=1)
```
