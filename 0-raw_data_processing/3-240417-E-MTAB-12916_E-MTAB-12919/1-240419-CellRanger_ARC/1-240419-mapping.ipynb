```python
import pandas as pd
```

```python
rna_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12916.sdrf.txt',
                                  index_col='Description')
rna_seq_meta_data.head()
```

```python
atac_seq_meta_data = pd.read_table('../.data/meta_data/E-MTAB-12919.sdrf.txt',
                                   index_col='Description')
atac_seq_meta_data.head()
```

```python
keys = ['rna', 'atac']
meta_data = pd.concat(
    [rna_seq_meta_data, atac_seq_meta_data],
    axis=1,
    keys= keys
)
meta_data.head()
```

```python
columns = [
    'Source Name',
    'Comment[SUBMITTED_FILE_NAME]',
    'Comment[FASTQ_URI]'
]

meta_data_download = pd.concat([meta_data[key][columns].add_suffix(f'_{key}') for key in keys], axis=1)
meta_data_download.head()
```

```python
meta_data_download.to_csv('../.data/meta_data/downloads_table.csv', header=None)
```

```python
%%bash

export PATH=/home/ceger/CellRanger/cellranger-arc-2.0.2:$PATH

metadata="/mnt/LaCIE/ceger/Projects/human_heart_mapping/human_heart_mapping/0-raw_data_processing/3-240417-E-MTAB-12916_E-MTAB-12919/.data/meta_data/downloads_table.csv"
output_path="/mnt/LaCIE/ceger/Projects/human_heart_mapping/human_heart_mapping/0-raw_data_processing/3-240417-E-MTAB-12916_E-MTAB-12919/.data/mapping"
index_output="/mnt/LaCIE/ceger/Projects/human_heart_mapping/human_heart_mapping/0-raw_data_processing/3-240417-E-MTAB-12916_E-MTAB-12919/.data/cr_arc_index/GRCh38"


if [ ! -f "$metadata" ]; then
    echo "File not found: $metadata"
    exit 1
fi

while IFS=, read -r index sample_rna file_name_rna url_rna sample_atac file_name_atac url_atac; do
    sample_output_path="$output_path/$index"
    rna_file_path="$sample_output_path/$sample_rna"
    atac_file_path="$sample_output_path/$sample_atac"
    rna_file="$rna_file_path/$file_name_rna"
    atac_file="$atac_file_path/$file_name_atac"

    mkdir -p $rna_file_path
    mkdir -p $atac_file_path

    echo "fastqs,sample,library_type" > "$sample_output_path/libraries.csv"
    echo "$rna_file_path","$sample_rna","Gene Expression" >> "$sample_output_path/libraries.csv"
    echo "$atac_file_path","$sample_atac","Chromatin Accessibility" >> "$sample_output_path/libraries.csv"


    axel -n 10 --output="$rna_file" "$url_rna"
    axel -n 10 --output="$atac_file" "$url_atac"

    if [ "$(find "$rna_file_path" -maxdepth 1 -type f | wc -l)" -eq 4 ] && [ "$(find "$atac_file_path" -maxdepth 1 -type f | wc -l)" -eq 4 ]; then
        echo "Both folders contain exactly 4 files."
        cellranger-arc count    --id="$index" \
                                --reference="/mnt/LaCIE/ceger/Projects/human_heart_mapping/human_heart_mapping/0-raw_data_processing/3-240417-E-MTAB-12916_E-MTAB-12919/.data/cr_arc_index/GRCh38" \
                                --libraries="/mnt/LaCIE/ceger/Projects/human_heart_mapping/human_heart_mapping/0-raw_data_processing/3-240417-E-MTAB-12916_E-MTAB-12919/.data/mapping/HCAHeart9508627_HCAHeart9508819/libraries.csv" \
                                --localcores=16 \
                                --localmem=32
        mv "$index" "$sample_output_path"
        rm -r "$rna_file_path"
        rm -r "$atac_file_path"
    fi
done < "$metadata"
```
